# My First Graph

We create a canvas holding the graphics primitives without relying on the default one root provides.

```python
c = ROOT.TCanvas()
c.SetGrid()
c.SetLogy()
```

```python
g = ROOT.TGraph()
for i in [0,1,2,3,4]:
    g.SetPoint(i,i,i*i)
```

Now the style

```python
g.SetMarkerStyle(ROOT.kFullTriangleUp)
g.SetMarkerSize(3)
g.SetMarkerColor(ROOT.kAzure)
g.SetLineColor(ROOT.kRed - 2)
g.SetLineWidth(11)
g.SetLineStyle(3)
g.SetTitle("My Graph;The X;My Y")
g.Draw("APL")
txt = "#color[804]{My text #mu {}^{40}_{20}Ca}"
l = ROOT.TLatex(.2, 10, txt)
l.Draw()
c.Draw()
```
